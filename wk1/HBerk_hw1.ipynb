```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
G=nx.florentine_families_graph()
plt.figure(3,figsize=(12,12))
labels=nx.draw_networkx_labels(G,pos=nx.spring_layout(G),font_size=14,font_family='Georgia',node_size=1000)
nx.draw(G)
```

```python
H=nx.krackhardt_kite_graph()
mapping={1:'Andre',2:'Beverley',3:'Carol',4:'Diane',5:'Ed',6:'Fernando',7:'Garth', 8:'Heather',9:'Ike',10:'Jane'}
H=nx.relabel_nodes(H,mapping)
plt.figure(3,figsize=(8,8))
pos=nx.drawing.spring_layout(H)
labels=nx.draw_networkx_labels(H,pos,font_size=12,font_family='Verdana', node_size=140)
nx.draw(H)
```
