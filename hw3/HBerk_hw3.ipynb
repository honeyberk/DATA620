```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate Star Wars Episode 4 (A New Hope) graph from graphml file
from networkx import read_graphml
K=read_graphml('star_wars_ep4.graphml')

plt.figure(figsize=(8,8))
pos=nx.drawing.spring_layout(K)
nx.draw(K, font_size=14,font_family='Verdana',font_weight='bold',node_size=800,node_color='blue',with_labels=True,alpha=0.5)
```

```python
# Diameter
K_diam = nx.diameter(K)
print 'The diameter of the Stars Wars Episode 4 graph is %s.' % K_diam
```

```python
# Average shortest path length
from networkx import algorithms

K_avg_shortest_path = algorithms.average_shortest_path_length(K)
print 'The average shortest path length of the Star Wars Episode 4 graph is %s.' % K_avg_shortest_path
```
